```
ranks.fillna(ranks.median(),inplace=True)
```

```
x = ranks.iloc[:, 4:8]
y = ranks.iloc[:, 3].values
```

```
from sklearn.preprocessing import MinMaxScaler, StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
y = y.reshape(-1,1)
y = sc.fit_transform(y)
```

```
from sklearn.model_selection import train_test_split
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```
from keras.models import Sequential
from keras.layers import Dense,Flatten
```

```
def model():
  ai = Sequential()

  ai.add(Dense(15,kernel_initializer='uniform', activation='relu', input_dim = 4))
  ai.add(Dense(8,kernel_initializer='uniform', activation="relu"))
  ai.add(Dense(1,kernel_initializer='uniform'))

  ai.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse','mae'])

  return ai
```

```
from keras.wrappers.scikit_learn import KerasRegressor
```

```
ai = KerasRegressor(build_fn=model, batch_size=5,epochs=100)
```

```
ai.fit(x_train,y_train)
```

```
y_pred = ai.predict(x_test)
```

```
Xnew = np.array([[40, 0, 26, 9000]])
Xnew= sc.transform(Xnew)
ynew= ai.predict(Xnew)
#invert normalize
ynew = sc.inverse_transform(ynew)
Xnew = sc.inverse_transform(Xnew)
print("X=%s, Predicted=%s" % (Xnew[0], ynew[0]))
```
